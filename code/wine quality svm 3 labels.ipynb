```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_auc_score
from sklearn import metrics
from sklearn.model_selection import GridSearchCV
from sklearn import preprocessing
wineQuality=pd.read_csv("E:\winequality-red.csv")
wineQuality
```

```python
wineQuality.describe()
```

```python
plt.subplots(figsize=(15, 10))
sns.heatmap(wineQuality.corr(), annot = True, cmap = 'coolwarm')
```

```python
wineQuality['label'] = ""
wineQuality['label'] = [2 if each > 6 else 1 if ((each > 4) and (each < 7)) else 0 for each in wineQuality['quality']]

print("Good Wine Class:",wineQuality[wineQuality['label'] == 2].shape)
print("Average Wine Class:",wineQuality[wineQuality['label'] == 1].shape)
print("Bad Wine Class:",wineQuality[wineQuality['label'] == 0].shape)
```

```python
fig, ax1 = plt.subplots(4,3, figsize=(22,16))
k = 0
for i in range(4):
    for j in range(3):
        if k != 11:
            sns.boxplot('label',wineQuality.iloc[:,k], data=wineQuality, ax = ax1[i][j])
            k += 1
plt.show()
```

```python
X3 = wineQuality.drop(['quality','label'], axis = 1)
y3 = wineQuality['label']
```

```python
X3_t = preprocessing.normalize (X3)
print("X_t:", X3_t.shape)
X3_train, X3_test, y3_train, y3_test = train_test_split(X3_t, y3, test_size=0.30, random_state=50)
print("X train:", X3_train.shape,"X test:",X3_test.shape)
```

```python
from sklearn.svm import SVC
svm3 = SVC(kernel='rbf',degree=3,C=1,probability=False,random_state=None,cache_size=200,max_iter=-1)
svm3.fit(X3_train, y3_train)
```

```python
print('Accuracy:',svm3.score(X3_test,y3_test))
```

```python
y_pred = svm3.predict(X3_test)
cnf_matrix = metrics.confusion_matrix(y3_test, y_pred)
cnf_matrix
```
