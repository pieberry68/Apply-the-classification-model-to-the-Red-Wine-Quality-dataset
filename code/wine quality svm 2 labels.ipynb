```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
wineQuality=pd.read_csv("E:\winequality-red.csv")
wineQuality
```

```python
from scipy import stats
z = np.abs(stats.zscore(wineQuality))
wineQuality = wineQuality[(z < 3).all(axis=1)]
wineQuality.shape
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.subplots(figsize=(15, 10))
sns.heatmap(wineQuality.corr(), annot = True, cmap = 'coolwarm')
```

```python
import numpy as np
wineQuality["isItGood"] =np.where(wineQuality["quality"] > 5, 1, 0)
#changing it to binary category
wineQuality
```

```python
features= ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol']
X = wineQuality[features] # Features
y = wineQuality.isItGood # Target variable
print(len(features))
```

```python
for i in range(11):
 sns.boxplot(x=y,y=X.iloc[:,i])
 plt.ylabel(features[i])
 plt.show()
```

```python
for i in range(11):
    plt.hist(X.iloc[:,i], edgecolor='white')
    plt.title(features[i])
    plt.show() 
```

```python
from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(X).transform(X)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=350) 
```

```python
kernel=['linear','rbf','poly']
for i in kernel:
    model=SVC(kernel=i,C=1.0)
    model.fit(X_train,y_train)
    print('Kernel:',i)
    print('Accuracy:',model.score(X_test,y_test))
```

```python
model=SVC()
model.fit(X_train,y_train)
print(" Accuracy on test data ",model.score(X_test,y_test))
print(" Accuracy on train data ",model.score(X_train,y_train))
```

```python
for i in range(1,5):
    model=SVC(kernel='poly',degree=i,C=100)
    model.fit(X_train,y_train)
    print(" Accuracy on test data ",model.score(X_test,y_test))
    print(" Accuracy on train data ",model.score(X_train,y_train))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={'C':[1,100,1000],'kernel':['linear','rbf','poly'],'degree':[1,2,3]}
grid=GridSearchCV(SVC(),param_grid)
grid.fit(X_train,y_train)
```

```python
print(grid.best_params_)
print(grid.score(X_test,y_test))
```
