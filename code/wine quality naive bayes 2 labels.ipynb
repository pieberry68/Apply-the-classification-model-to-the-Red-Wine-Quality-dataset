```python
import pandas as pd
import numpy as np
wineQuality=pd.read_csv("E:\winequality-red.csv")
import warnings
warnings.filterwarnings('ignore')
wineQuality
```

```python
wineQuality.describe()
```

```python
from scipy import stats
z = np.abs(stats.zscore(wineQuality))
wineQuality = wineQuality[(z < 3).all(axis=1)]
wineQuality.shape
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.subplots(figsize=(15, 10))
sns.heatmap(wineQuality.corr(), annot = True, cmap = 'coolwarm')
```

```python
wineQuality["isItGood"] =np.where(wineQuality["quality"] > 5, 1, 0)
#changing it to binary category
wineQuality
```

```python
feature_cols = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol']
X = wineQuality[feature_cols] # Features
y = wineQuality.isItGood # Target variable
```

```python
from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(X).transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=350) 
X_train
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)
```

```python
y_pred
```

```python
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
print(metrics.classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import RepeatedStratifiedKFold
cv_method = RepeatedStratifiedKFold(n_splits=5,  n_repeats=3, random_state=999)
from sklearn.preprocessing import PowerTransformer
params_NB = {'var_smoothing':np.logspace(0,-9, num=100)}
from sklearn.model_selection import GridSearchCV
gs_NB = GridSearchCV(estimator=GaussianNB(), param_grid=params_NB, cv=cv_method,verbose=1,scoring='accuracy')
Data_transformed = PowerTransformer().fit_transform(X_test)
gs_NB.fit(Data_transformed, y_test);
```

```python
results_NB = pd.DataFrame(gs_NB.cv_results_['params'])
```

```python
results_NB['test_score'] = gs_NB.cv_results_['mean_test_score']
predict_test = gs_NB.predict(Data_transformed)
```

```python
from sklearn import metrics
print('accuracy_score on test dataset : ', metrics.accuracy_score(y_test,predict_test))
```
